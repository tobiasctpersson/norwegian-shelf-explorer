## Import

```python
import pandas as pd

url = "https://factpages.sodir.no/en/wellbore/tableview/exploration/all"
df = pd.read_html(url)[0]
df['Entered date'] = pd.to_datetime(df['Entered date'], dayfirst=True)
print(f"Loaded {len(df)} wells")
```

## Insight 1 - Who are Wellvector's best sales targets?

```python
operator_archives = df[df['Status'] == 'P&A']['Drilling operator'].value_counts().head(10)
print("Operators with most P&A wells (biggest archives):")
print(operator_archives)
```

## Insight 2 - How much data was released via reliquishment?

```python
released = df[df['Release date'].notna()]
print(f"Wells with public release date: {len(released)}")
print(f"That's {len(released)/len(df)*100:.1f}% of all exploration wells")
```

## Insight 3 - Drilling activity over time

```python
df['Year'] = df['Entered date'].dt.year
yearly = df.groupby('Year').size()
print("Wells drilled per year:")
print(yearly)
```
