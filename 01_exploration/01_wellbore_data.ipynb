```python
import pandas as pd

url = "https://factpages.sodir.no/en/wellbore/tableview/exploration/all"
tables = pd.read_html(url)
df = tables[0]
print(f"Total exploration wells: {len(df)}")
df.head()
```

## What columns do we have?

```python
print(df.columns.tolist())
```

## Which operators have drilled the most?

```python
df['Drilling operator'].value_counts().head(10)
```

## Breakdown by content (oil, gas, dry...)

```python
df['Content'].value_counts()
```

## Breakdown by status (how many are P&A'd = plugged and abandoned?)

```python
df['Status'].value_counts()
```

## How many wells per decade?

```python
df['Entered date'] = pd.to_datetime(df['Entered date'], dayfirst=True)
df['Decade'] = (df['Entered date'].dt.year // 10 * 10).astype(str) + 's'
df['Decade'].value_counts().sort_index()
```

## Information

```python
pa_wells = df[df['Status'] == 'P&A']
print(f"P&A wells: {len(pa_wells)}")
print(f"Percentage of total: {len(pa_wells)/len(df)*100:.1f}%")
print(f"Avg drilling days: {df['Drilling days'].mean():.0f} days per well")
print(f"Date range: {df['Entered date'].min().year} to {df['Entered date'].max().year}")
```
